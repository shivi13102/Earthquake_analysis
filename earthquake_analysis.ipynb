# Cleaning df1 Table

```python
columns_to_check = ['date_time','latitude', 'longitude', 'location', 'continent', 'country']

# Check for null values in df1
null_values_df1 = df1[columns_to_check].isnull().sum()

print("Null values in df1:")
print(null_values_df1)
```

```python
display(df1.head(10))
```

```python
# Use a regular expression to split 'location' into 'location' and 'country' only if a comma exists
df1[['location', 'country']] = df1['location'].str.extract(r'^\s*([^,]+)\s*(?:,\s*(.*))?$', expand=True)

# Fill empty 'country' values with an empty string
df1['country'] = df1['country'].fillna("")

# Verify the changes
display(df1.head(10))
```

```python
# Delete rows where 'location' column is null in both dataframes
df1 = df1[df1['location'].notnull()]

# Verify by checking the number of null values again
null_values_df1_cleaned = df1['location'].isnull().sum()

# Print the cleaned dataframes' summary
print(f"Null values in 'location' for df1 after cleaning: {null_values_df1_cleaned}")
```

```python
display(df1)
```

```python
# Reorder columns in df1
columns = list(df1.columns)
columns.remove('country')
columns.insert(columns.index('location') + 1, 'country')
df1 = df1[columns]

display(df1.head())
```

```python
columns_to_check = ['title', 'magnitude', 'date_time', 'cdi', 'mmi', 'alert', 'tsunami', 'sig', 'net', 'nst', 'dmin', 'gap', 'magType', 'depth']

# Check for null values in df1
null_values_df1 = df1[columns_to_check].isnull().sum()

print("Null values in df1:")
print(null_values_df1)
```
