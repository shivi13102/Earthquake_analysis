# Cleaning df2 Table

```python
df3 = pd.read_csv("/content/cleaned_earthquake_data.csv")
display(df3)
```

```python
print(df3.columns)
print(df2.columns)
```

```python
# Insert 'alert' column from df2 into df3 after 'mmi'
mmi_index = df3.columns.get_loc('mmi')  # Get the index of the 'mmi' column
df3.insert(mmi_index + 1, 'alert', df2['alert'])

display(df3)
```

```python
# Fix date parsing issue
df5['date_time'] = pd.to_datetime(df5['date_time'], errors='coerce', dayfirst=True)
df3['date_time'] = pd.to_datetime(df3['date_time'], errors='coerce', dayfirst=True)

# Convert all int columns in df1 to float
int_columns = df5.select_dtypes(include=['int64', 'int32']).columns
df5[int_columns] = df5[int_columns].astype(float)

# Check for common rows
common_rows = pd.merge(df5, df3, how='inner')

if not common_rows.empty:
    print("Common rows:")
    print(common_rows)
else:
    print("No common rows found.")
```

# Final merging of both the tables (Normalized)

```python
# Merge df1 below df3 (concatenate the tables)
merged_df1 = pd.concat([df3, df5], ignore_index=True)

# Sort the final table by 'date_time' in descending order
merged_df1 = merged_df1.sort_values(by='date_time', ascending=False)

# Reset the index for a clean DataFrame
merged_df1 = merged_df1.reset_index(drop=True)

display(merged_df1)
```
